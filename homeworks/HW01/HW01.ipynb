```python
import numpy as np
#Создадим одномерный массив из списка
a1=np.array([1,3,5,7,9,11,13])
a1
```

```python
#Теперь создадим двумерный массив
a2=np.array([[1,3,5,7],[2,4,6,8]])
a2
```

```python
#Создадим двумерный массив из нулей, размером 5 на 5
a3=np.zeros((5,5))
a3
```

```python
#Создадим двумерный массив из нулей, размером 2 на 4
a4=np.ones((2,4))
a4
```

```python
#Создадим массив состоящий из чётных числе в диапазоне от 2 до 14
a5=np.arange(2,15,2)
a5
```

```python
#Выведем атрибуты для массива a5
print("Форма массива:",a5.shape)
print("Тип массива:", a5.dtype)
```

```python
#С помощью индексов получим элемент массива a2 на позции второй строки и 3 столбца
a2[1,2]
```

```python
#С помощью срезов получим элементы массива a1 с пятого по шестой
a1[4:6]
```

```python
#Поэлементно сложим массивы a1 и a5
a1+a5
```

```python
#Поэлементно умножаем массивы a1 и a5
a1*a5
```

```python
#Вычисляем среднее значение элементов массива a2
np.mean(a2)
```
